```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
positive_data = pd.read_csv("data/training/positive/positive_data.csv")
negative_data = pd.read_csv("data/training/negative/negative_data.csv")
```

```python


plt.figure(figsize=(10, 6))
plt.hist(negative_data['cosine_similarity'], bins=1000, color='red', alpha=0.5, label='Negative')
plt.hist(positive_data['cosine_similarity'], bins=1000, color='blue', alpha=0.5, label='Positive')
plt.title('Distribution of Cosine Similarities')
plt.xlabel('Cosine Similarity')
plt.ylabel('Count')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(negative_data['euclidean_distance'], bins=1000, color='red', alpha=0.5, label='Negative')
plt.hist(positive_data['euclidean_distance'], bins=1000, color='blue', alpha=0.5, label='Positive')
plt.title('Distribution of Euclidean Distances')
plt.xlabel('Euclidean Distance')
plt.ylabel('Count')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(negative_data['dot_product'], bins=1000, color='red', alpha=0.5, label='Negative')
plt.hist(positive_data['dot_product'], bins=1000, color='blue', alpha=0.5, label='Positive')
plt.title('Distribution of Dot Products')
plt.xlabel('Dot Product')
plt.ylabel('Count')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
