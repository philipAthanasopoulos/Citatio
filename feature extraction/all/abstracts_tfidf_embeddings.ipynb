```python
import pandas as pd

abstracts = pd.read_csv('../../data/abstracts.txt', sep=r'\|\-\-\|', header=None, index_col=False, engine='python')
abstracts.columns = ['paper_id', 'abstract']

abstracts['abstract_length'] = abstracts['abstract'].fillna('').apply(lambda x: len(x.split()))
from sklearn.feature_extraction.text import TfidfVectorizer

# Fill missing abstracts with empty strings
abstracts['abstract'] = abstracts['abstract'].fillna('')

# Initialize TF-IDF Vectorizer
tfidf = TfidfVectorizer(max_features=300, stop_words='english')  # You can adjust max_features as needed

# Fit and transform abstracts to get TF-IDF embeddings
tfidf_matrix = tfidf.fit_transform(abstracts['abstract'])

# Convert the TF-IDF matrix to a DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
tfidf_df['paper_id'] = abstracts['paper_id'].values

tfidf_df.head()

# Save the embeddings to a CSV file
tfidf_df.to_csv('../../data/abstracts_tfidf_embeddings.csv', index=False)
```
