```python
import pandas as pd
import numpy as np
from sklearn.metrics import jaccard_score
from tqdm.notebook import tqdm

negative_edges = pd.read_csv('../../data/training/negative/non_existing.csv')
authors = pd.read_csv('../../data/authors.txt', sep=r'\|\-\-\|', header=None, index_col=False, engine='python')

negative_edges['source_authors'] = authors.loc[negative_edges['source'], 1].values
negative_edges['target_authors'] = authors.loc[negative_edges['target'], 1].values

jaccard_scores = []
for i in tqdm(range(len(negative_edges))):
    source_authors = set(str(negative_edges['source_authors'].iloc[i]).split(','))
    target_authors = set(str(negative_edges['target_authors'].iloc[i]).split(','))
    intersection = len(source_authors & target_authors)
    union = len(source_authors | target_authors)
    score = intersection / union if union > 0 else 0.0
    jaccard_scores.append(score)
negative_edges['jaccard_score'] = jaccard_scores
negative_edges = negative_edges[['source', 'target', 'jaccard_score']]
negative_edges.head()
negative_edges.to_csv('../../data/training/negative/negative_edges_authors_jaccard.csv', index=False)

count_greater_than_zero = (negative_edges['jaccard_score'] > 0).sum()
print(count_greater_than_zero)
```
