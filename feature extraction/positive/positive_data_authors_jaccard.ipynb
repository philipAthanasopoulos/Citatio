```python
import pandas as pd
import numpy as np
from sklearn.metrics import jaccard_score
from tqdm.notebook import tqdm

positive_edges = pd.read_csv('../../data/edgelist.txt', sep=',', header=None)
positive_edges.columns = ['source', 'target']
authors = pd.read_csv('../../data/authors.txt', sep=r'\|\-\-\|', header=None, index_col=False, engine='python')

positive_edges['source_authors'] = authors.loc[positive_edges['source'], 1].values
positive_edges['target_authors'] = authors.loc[positive_edges['target'], 1].values

jaccard_scores = []
for i in tqdm(range(len(positive_edges))):
    source_authors = set(str(positive_edges['source_authors'].iloc[i]).split(','))
    target_authors = set(str(positive_edges['target_authors'].iloc[i]).split(','))
    intersection = len(source_authors & target_authors)
    union = len(source_authors | target_authors)
    score = intersection / union if union > 0 else 0.0
    jaccard_scores.append(score)
positive_edges['jaccard_score'] = jaccard_scores
positive_edges = positive_edges[['source', 'target', 'jaccard_score']]
positive_edges.head()
positive_edges.to_csv('../../data/training/positive/positive_edges_authors_jaccard.csv', index=False)

count_greater_than_zero = (positive_edges['jaccard_score'] > 0).sum()
print(count_greater_than_zero)
```
