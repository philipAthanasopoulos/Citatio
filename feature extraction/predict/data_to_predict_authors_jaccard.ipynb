```python
import pandas as pd
import numpy as np
from sklearn.metrics import jaccard_score
from tqdm.notebook import tqdm

predict_edges = pd.read_csv('../../data/test.txt', sep=',', header=None)
predict_edges.columns = ['source', 'target']
authors = pd.read_csv('../../data/authors.txt', sep=r'\|\-\-\|', header=None, index_col=False, engine='python')


predict_edges['source_authors'] = authors.loc[predict_edges['source'], 1].values
predict_edges['target_authors'] = authors.loc[predict_edges['target'], 1].values

jaccard_scores = []
for i in tqdm(range(len(predict_edges))):
    source_authors = set(str(predict_edges['source_authors'].iloc[i]).split(','))
    target_authors = set(str(predict_edges['target_authors'].iloc[i]).split(','))
    intersection = len(source_authors & target_authors)
    union = len(source_authors | target_authors)
    score = intersection / union if union > 0 else 0.0
    jaccard_scores.append(score)
predict_edges['jaccard_score'] = jaccard_scores
predict_edges = predict_edges[['source', 'target', 'jaccard_score']]
predict_edges.head()
predict_edges.to_csv('../../data/training/predict/predict_edges_authors_jaccard.csv', index=False)

count_greater_than_zero = (predict_edges['jaccard_score'] > 0).sum()
print(count_greater_than_zero)
```
